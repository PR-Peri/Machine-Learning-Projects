```python
import pandas as pd

xdf = pd.read_csv('xiaomi_file1.csv')  
xdf.to_csv('sentiment_of_xiaomi.csv', header = ['Tweets'])
```

```python
xdf1 = pd.read_csv('sentiment_of_xiaomi.csv') 
xdf1.shape
```

```python
import re
import string

def clean_text(text):
  text = re.sub('@[\w]*','',text)
  text = re.sub('\[.*?\]','',text)
  text = re.sub('[%s]'% re.escape(string.punctuation), '',text)
  text = re.sub('\w*\d\w*','',text)
  text= re.sub(r'@[A-Za-z0-9]+','',text) #remove mention
  text = re.sub('https?://[A-Za-z0-9./]+','',text) #remove https
  text = re.sub('b\w*','',text)
  return text

cleaned = lambda x: clean_text(x)
```

```python
xdf1['Tweets']= pd.DataFrame(xdf1.Tweets.apply(cleaned))
xdf1
```

```python
xdf2=xdf1.drop(["Unnamed: 0"], axis=1)
xdf2
```

```python
from textblob import TextBlob
import nltk
nltk.download('punkt')
nltk.download('brown')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
def get_polarity(text):
  return TextBlob(text).sentiment.polarity

xdf2['TBpolarity'] = xdf2['Tweets'].apply(get_polarity)
```

```python
xdf3=xdf2.drop_duplicates()
```

```python
xdf3['Sentiment_type']= ''
xdf3.loc[xdf3.TBpolarity>0, 'Sentiment_type']='POSITIVE'
xdf3.loc[xdf3.TBpolarity==0, 'Sentiment_type']='NEUTRAL'
xdf3.loc[xdf3.TBpolarity<0, 'Sentiment_type']='NEGATIVE'
xdf4=xdf3.head(200)
```
