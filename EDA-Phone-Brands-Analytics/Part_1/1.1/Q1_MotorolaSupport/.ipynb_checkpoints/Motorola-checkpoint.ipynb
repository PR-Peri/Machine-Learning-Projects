```python
import csv
csvFile = open('moto_file1.csv', 'w')
counter=0

csvWriter = csv.writer(csvFile)

for tweet in tweepy.Cursor(api.search,q='#MotoG10Power',count=4000,lang="en", tweet_mode='extended').items(4000):
    print (tweet.full_text)
    csvWriter.writerow([tweet.full_text.encode('utf-8')])
counter+=1
print("Total %d User Timeline Tweets collected" % counter)
csvFile.close()
```

```python
mdf1 = pd.read_csv('sentiment_of_motorola.csv') 
mdf1.shape
```

```python
import re
import string

def clean_text(text):
  text = re.sub('@[\w]*','',text)
  text = re.sub('\[.*?\]','',text)
  text = re.sub('[%s]'% re.escape(string.punctuation), '',text)
  text = re.sub('\w*\d\w*','',text)
  text= re.sub(r'@[A-Za-z0-9]+','',text) #remove mention
  text = re.sub('https?://[A-Za-z0-9./]+','',text) #remove https
  text = re.sub('b\w*','',text)
  return text

cleaned = lambda x: clean_text(x)
```

```python
mdf1['Tweets']= pd.DataFrame(mdf1.Tweets.apply(cleaned))
mdf1
```

```python
mdf2=mdf1.drop(["Unnamed: 0"], axis=1)
mdf2
```

```python
mdf3=mdf2.drop_duplicates()
mdf3.shape
```

```python
from textblob import TextBlob
import nltk
nltk.download('punkt')
nltk.download('brown')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
def get_polarity(text):
  return TextBlob(text).sentiment.polarity

mdf3['TBpolarity'] = mdf3['Tweets'].apply(get_polarity)
```

```python
md4 = mdf3.head(200)
md4
```

```python
md4['Sentiment_type']= ''
md4.loc[md4.TBpolarity>0, 'Sentiment_type']='POSITIVE'
md4.loc[md4.TBpolarity==0, 'Sentiment_type']='NEUTRAL '
md4.loc[md4.TBpolarity<0, 'Sentiment_type']='NEGATIVE'
md4.head()
```
