### LICENSE PLATE EVALUATION FOR KAGGLE DATASET

```python
import pandas as pd
import numpy as np
```

```python
header = ["Plate"]
dt=pd.read_csv('groundtruth.csv',names=header)
dt.dtypes
```

```python
dt['Plate'].str.split('',expand=True).head(5)
# dt
```

```python
header2 = ["Plate"]
ANN_df=pd.read_csv('ANN_output.csv',skip_blank_lines=False,names=header2)
ANN_df.dtypes
```

```python
ANN_df['Plate'].str.split('',expand=True)
```

```python
cp = pd.merge(dt, ANN_df, how='inner', on=['Plate']).count()
cp
```

```python
missing = dt['Plate'].count() - ANN_df['Plate'].count()
missing
```

```python
fail = 230 - cp - missing
fail
```

```python
correct = cp/230
miss = missing/230
fail_detect = fail/230
```

```python
print ('Correct / Overall Plates -> ' , round(correct, 2).to_string())
print ('Missing / Overall Plates -> ' , round(miss, 2))
print ('Incorrect / Overall Plates -> ' , round(fail_detect, 2).to_string())
```

```python
Accuracy = correct * 100
print('Accuracy', round(Accuracy, 2).to_string(), '%')
```
