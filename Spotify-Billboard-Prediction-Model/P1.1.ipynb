```python
data = pd.read_csv("final.csv")
```

```python
data["WeekID"]= pd.to_datetime(data["WeekID"]) 
data.info()
```

```python
def freq(col):
    g= sns.kdeplot((data[col][(data["WeekID"] == "1959-05-16")]) , color = "Blue", shade = False, label = "1959")
    g= sns.kdeplot((data[col][(data["WeekID"] == "2019-04-06")]) , color = "Red", shade = False, label = "2019")
    g.set_xlabel(col)
    g.set_ylabel("Frequency")
    plt.show()
```
